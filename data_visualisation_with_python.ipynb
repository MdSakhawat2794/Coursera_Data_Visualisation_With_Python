**Task 1.2 - Plot different lines for categories of vehicle type and analyse the trend to answer the question "Is there a noticeable difference in sales trends between different vehicle types during recession periods?"**



```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data creation
# Replace this with the actual data loading step
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020] * 3,
    'Sales': [150, 180, 220, 300, 330, 360, 390, 400, 420, 410, 380] +
             [100, 120, 180, 200, 220, 250, 270, 290, 310, 300, 280] +
             [200, 220, 240, 260, 280, 300, 320, 340, 360, 350, 340],
    'Vehicle_Type': ['Sedan'] * 11 + ['SUV'] * 11 + ['Truck'] * 11,
    'Recession': [True, True, False, False, False, True, True, False, False, True, True] * 3
})

# Filter data for recession periods
recession_data = data[data['Recession'] == True]

# Plotting the data
plt.figure(figsize=(10, 6))

# Loop through each vehicle type and plot the data
for vehicle_type in recession_data['Vehicle_Type'].unique():
    subset = recession_data[recession_data['Vehicle_Type'] == vehicle_type]
    plt.plot(subset['Year'], subset['Sales'], marker='o', linestyle='-', label=vehicle_type)

plt.title('Sales Trends by Vehicle Type During Recession Periods')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(title='Vehicle Type')
plt.grid(True)
plt.xticks(recession_data['Year'].unique())
plt.show()
```

**Task 1.4 - Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data creation
# Replace this with the actual data loading step
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'GDP': [14.4, 14.6, 14.8, 15.1, 15.5, 15.8, 16.1, 16.3, 16.6, 16.8, 17.1],
    'Recession': [True, True, False, False, False, True, True, False, False, True, True]
})

# Filter data for recession and non-recession periods
recession_data = data[data['Recession'] == True]
non_recession_data = data[data['Recession'] == False]

# Creating subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Plotting GDP during recession periods
axs[0].plot(recession_data['Year'], recession_data['GDP'], marker='o', linestyle='-', color='r')
axs[0].set_title('GDP During Recession Periods')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('GDP')
axs[0].grid(True)

# Plotting GDP during non-recession periods
axs[1].plot(non_recession_data['Year'], non_recession_data['GDP'], marker='o', linestyle='-', color='g')
axs[1].set_title('GDP During Non-Recession Periods')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('GDP')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

**Task 1.5 - Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data creation
# Replace this with the actual data loading step
data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [120, 150, 180, 220, 210, 190, 170, 160, 200, 230, 240, 220],
    'Season': ['Winter', 'Winter', 'Spring', 'Spring', 'Spring', 'Summer', 'Summer', 'Summer', 'Autumn', 'Autumn', 'Autumn', 'Winter']
})

# Calculate the size of bubbles (proportional to sales)
data['BubbleSize'] = data['Sales'] * 10  # Adjust this multiplier as needed

# Create the bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(data['Month'], data['Sales'], s=data['BubbleSize'], alpha=0.5, c='b', edgecolors='w', linewidth=0.5)

plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)

# Optionally, annotate each bubble with the corresponding sales value
for i in range(len(data)):
    plt.text(data['Month'][i], data['Sales'][i], data['Sales'][i], ha='center', va='center', fontsize=9, color='black')

plt.show()
```

**Task 1.7 - Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data creation
# Replace this with the actual data loading step
data = pd.DataFrame({
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020] * 2,
    'Ad_Expenditure': [100, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210] +
                      [110, 130, 120, 150, 140, 170, 160, 190, 180, 210, 200],
    'Recession': [True, True, False, False, False, True, True, False, False, True, True] * 2
})

# Aggregate data by recession status
ad_exp_recession = data[data['Recession'] == True]['Ad_Expenditure'].sum()
ad_exp_non_recession = data[data['Recession'] == False]['Ad_Expenditure'].sum()

# Create a pie chart
labels = ['Recession', 'Non-Recession']
sizes = [ad_exp_recession, ad_exp_non_recession]
colors = ['red', 'green']
explode = (0.1, 0)  # Explode the first slice (Recession)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title('Advertising Expenditure of XYZAutomotives During Recession and Non-Recession Periods')
plt.show()
```

**Task 2.4 - Creating Callbacks; Define the callback function to update the input container based on the selected statistics and the output container.**

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

# Initialize the Dash app
app = dash.Dash(__name__)

# Give the application a meaningful title
app.title = "Automobile Sales Analysis Dashboard"

# Define the drop-down options
vehicle_type_options = [
    {'label': 'Sedan', 'value': 'Sedan'},
    {'label': 'SUV', 'value': 'SUV'},
    {'label': 'Truck', 'value': 'Truck'}
]

year_options = [
    {'label': str(year), 'value': year} for year in range(2010, 2021)
]

# Define the layout of the app
app.layout = html.Div(children=[
    html.H1(children='Automobile Sales Analysis Dashboard'),

    html.Div(children='''
        A web application for analyzing automobile sales data.
    '''),

    html.Label('Select Vehicle Type:'),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=vehicle_type_options,
        value='Sedan',  # Default value
        clearable=False
    ),

    html.Label('Select Year:'),
    dcc.Dropdown(
        id='year-dropdown',
        options=year_options,
        value=2010,  # Default value
        clearable=False
    ),

    html.Div(id='output-container', className='output-display'),

    dcc.Graph(id='sales-graph')
])

# Define the callback to update the output container
@app.callback(
    Output('output-container', 'children'),
    Output('sales-graph', 'figure'),
    Input('vehicle-type-dropdown', 'value'),
    Input('year-dropdown', 'value')
)
def update_output(vehicle_type, year):
    # Dummy data for the example
    sales_data = {
        'Sedan': [200, 210, 220, 230, 240, 250, 260, 270, 280, 290, 300],
        'SUV': [150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250],
        'Truck': [100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200]
    }

    years = list(range(2010, 2021))
    selected_sales = sales_data.get(vehicle_type, [])

    # Update output container
    output_message = f'Selected Vehicle Type: {vehicle_type}, Year: {year}'

    # Update graph
    figure = {
        'data': [
            {'x': years, 'y': selected_sales, 'type': 'line', 'name': vehicle_type}
        ],
        'layout': {
            'title': f'Sales Data for {vehicle_type} ({year})',
            'xaxis': {'title': 'Year'},
            'yaxis': {'title': 'Sales'}
        }
    }

    return output_message, figure

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

**Task 2.5 - Create and display graphs for Recession Report Statistics.**

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd

# Initialize the Dash app
app = dash.Dash(__name__)

# Give the application a meaningful title
app.title = "Automobile Sales Analysis Dashboard"

# Example data creation
data = pd.DataFrame({
    'Year': list(range(2010, 2021)) * 3,
    'Sales': [150, 180, 220, 300, 330, 360, 390, 400, 420, 410, 380] * 3,
    'Average_Price': [20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000, 30000] * 3,
    'Unemployment_Rate': [7.5, 7.8, 7.9, 8.0, 7.7, 7.5, 7.3, 7.2, 7.0, 6.8, 6.7] * 3,
    'Vehicle_Type': ['Sedan', 'SUV', 'Truck'] * 11,
    'Recession': [True, True, False, False, False, True, True, False, False, True, True] * 3
})

# Filter data for recession periods
recession_data = data[data['Recession'] == True]

# Define the drop-down options
vehicle_type_options = [
    {'label': 'Sedan', 'value': 'Sedan'},
    {'label': 'SUV', 'value': 'SUV'},
    {'label': 'Truck', 'value': 'Truck'}
]

# Define the layout of the app
app.layout = html.Div(children=[
    html.H1(children='Automobile Sales Analysis Dashboard'),

    html.Div(children='''
        A web application for analyzing automobile sales data.
    '''),

    html.Label('Select Vehicle Type:'),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=vehicle_type_options,
        value='Sedan',  # Default value
        clearable=False
    ),

    html.Div(id='output-container', className='output-display'),

    dcc.Graph(id='sales-graph'),
    dcc.Graph(id='average-price-graph'),
    dcc.Graph(id='unemployment-rate-graph')
])

# Define the callback to update the output container and graphs
@app.callback(
    [Output('output-container', 'children'),
     Output('sales-graph', 'figure'),
     Output('average-price-graph', 'figure'),
     Output('unemployment-rate-graph', 'figure')],
    [Input('vehicle-type-dropdown', 'value')]
)
def update_output(vehicle_type):
    filtered_data = recession_data[recession_data['Vehicle_Type'] == vehicle_type]

    # Update output container
    output_message = f'Selected Vehicle Type: {vehicle_type}'

    # Update sales graph
    sales_figure = {
        'data': [
            go.Scatter(
                x=filtered_data['Year'],
                y=filtered_data['Sales'],
                mode='lines+markers',
                name='Sales'
            )
        ],
        'layout': go.Layout(
            title=f'Sales Data for {vehicle_type} During Recession',
            xaxis={'title': 'Year'},
            yaxis={'title': 'Sales'}
        )
    }

    # Update average price graph
    average_price_figure = {
        'data': [
            go.Scatter(
                x=filtered_data['Year'],
                y=filtered_data['Average_Price'],
                mode='lines+markers',
                name='Average Price'
            )
        ],
        'layout': go.Layout(
            title=f'Average Price Data for {vehicle_type} During Recession',
            xaxis={'title': 'Year'},
            yaxis={'title': 'Average Price'}
        )
    }

    # Update unemployment rate graph
    unemployment_rate_figure = {
        'data': [
            go.Scatter(
                x=filtered_data['Year'],
                y=filtered_data['Unemployment_Rate'],
                mode='lines+markers',
                name='Unemployment Rate'
            )
        ],
        'layout': go.Layout(
            title=f'Unemployment Rate During Recession',
            xaxis={'title': 'Year'},
            yaxis={'title': 'Unemployment Rate'}
        )
    }

    return output_message, sales_figure, average_price_figure, unemployment_rate_figure

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd

# Initialize the Dash app
app = dash.Dash(__name__)

# Give the application a meaningful title
app.title = "Automobile Sales Analysis Dashboard"

# Example data creation
data = pd.DataFrame({
    'Year': list(range(2010, 2021)) * 3,
    'Sales': [150, 180, 220, 300, 330, 360, 390, 400, 420, 410, 380] * 3,
    'Average_Price': [20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000, 30000] * 3,
    'Unemployment_Rate': [7.5, 7.8, 7.9, 8.0, 7.7, 7.5, 7.3, 7.2, 7.0, 6.8, 6.7] * 3,
    'Vehicle_Type': ['Sedan', 'SUV', 'Truck'] * 11,
    'Recession': [True, True, False, False, False, True, True, False, False, True, True] * 3
})

# Define the drop-down options
vehicle_type_options = [
    {'label': 'Sedan', 'value': 'Sedan'},
    {'label': 'SUV', 'value': 'SUV'},
    {'label': 'Truck', 'value': 'Truck'}
]

year_options = [
    {'label': str(year), 'value': year} for year in range(2010, 2021)
]

# Define the layout of the app
app.layout = html.Div(children=[
    html.H1(children='Automobile Sales Analysis Dashboard'),

    html.Div(children='''
        A web application for analyzing automobile sales data.
    '''),

    html.Label('Select Vehicle Type:'),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=vehicle_type_options,
        value='Sedan',  # Default value
        clearable=False
    ),

    html.Label('Select Year:'),
    dcc.Dropdown(
        id='year-dropdown',
        options=year_options,
        value=2010,  # Default value
        clearable=False
    ),

    html.Div(id='output-container', className='output-display'),

    dcc.Graph(id='sales-graph'),
    dcc.Graph(id='average-price-graph'),
    dcc.Graph(id='unemployment-rate-graph')
])

# Define the callback to update the output container and graphs for recession report statistics
@app.callback(
    [Output('output-container', 'children'),
     Output('sales-graph', 'figure'),
     Output('average-price-graph', 'figure'),
     Output('unemployment-rate-graph', 'figure')],
    [Input('vehicle-type-dropdown', 'value')]
)
def update_recession_report(vehicle_type):
    filtered_data = data[(data['Vehicle_Type'] == vehicle_type) & (data['Recession'] == True)]

    # Update output container
    output_message = f'Selected Vehicle Type: {vehicle_type} (Recession Period)'

    # Update sales graph
    sales_figure = {
        'data': [
            go.Scatter(
                x=filtered_data['Year'],
                y=filtered_data['Sales'],
                mode='lines+markers',
                name='Sales'
            )
        ],
        'layout': go.Layout(
            title=f'Sales Data for {vehicle_type} During Recession',
            xaxis={'title': 'Year'},
            yaxis={'title': 'Sales'}
        )
    }

    # Update average price graph
    average_price_figure = {
        'data': [
            go.Scatter(
                x=filtered_data['Year'],
                y=filtered_data['Average_Price'],
                mode='lines+markers',
                name='Average Price'
            )
        ],
        'layout': go.Layout(
            title=f'Average Price Data for {vehicle_type} During Recession',
            xaxis={'title': 'Year'},
            yaxis={'title': 'Average Price'}
        )
    }

    # Update unemployment rate graph
    unemployment_rate_figure = {
        'data': [
            go.Scatter(
                x=filtered_data['Year'],
                y=filtered_data['Unemployment_Rate'],
                mode='lines+markers',
                name='Unemployment Rate'
            )
        ],
        'layout': go.Layout(
            title=f'Unemployment Rate During Recession',
            xaxis={'title': 'Year'},
            yaxis={'title': 'Unemployment Rate'}
        )
    }

    return output_message, sales_figure, average_price_figure, unemployment_rate_figure

# Define the callback to update the graphs for yearly report statistics
@app.callback(
    [Output('sales-graph', 'figure'),
     Output('average-price-graph', 'figure'),
     Output('unemployment-rate-graph', 'figure')],
    [Input('year-dropdown', 'value')]
)
def update_yearly_report(year):
    filtered_data = data[data['Year'] == year]

    # Update sales graph
    sales_figure = {
        'data': [
            go.Bar(
                x=filtered_data['Vehicle_Type'],
                y=filtered_data['Sales'],
                name='Sales'
            )
        ],
        'layout': go.Layout(
            title=f'Sales Data for {year}',
            xaxis={'title': 'Vehicle Type'},
            yaxis={'title': 'Sales'}
        )
    }

    # Update average price graph
    average_price_figure = {
        'data': [
            go.Bar(
                x=filtered_data['Vehicle_Type'],
                y=filtered_data['Average_Price'],
                name='Average Price'
            )
        ],
        'layout': go.Layout(
            title=f'Average Price Data for {year}',
            xaxis={'title': 'Vehicle Type'},
            yaxis={'title': 'Average Price'}
        )
    }

    # Update unemployment rate graph
    unemployment_rate_figure = {
        'data': [
            go.Bar(
                x=filtered_data['Vehicle_Type'],
                y=filtered_data['Unemployment_Rate'],
                name='Unemployment Rate'
            )
        ],
        'layout': go.Layout(
            title=f'Unemployment Rate for {year}',
            xaxis={'title': 'Vehicle Type'},
            yaxis={'title': 'Unemployment Rate'}
        )
    }

    return sales_figure, average_price_figure, unemployment_rate_figure

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
